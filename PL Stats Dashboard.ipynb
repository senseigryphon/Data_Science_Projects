```python
import dash
from dash import dcc, html
import plotly.express as px
from dash.dependencies import Input, Output
```

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1('Premier League Team Statistical Comparison'),
    dcc.Dropdown(
        id='x-axis-dropdown',
        options=[{'label': col, 'value': col} for col in merged_df.columns if col != 'Squad'],
        value='xGD',  # Default value
        clearable=False,
        style={"width": "50%"}
    ),
    dcc.Dropdown(
        id='y-axis-dropdown',
        options=[{'label': col, 'value': col} for col in merged_df.columns if col != 'Squad'],
        value='Points Per Match',  # Default value
        clearable=False,
        style={"width": "50%"}
    ),
    dcc.Graph(id='scatter-plot')
])

# Callback to update the graph based on dropdowns
@app.callback(
    Output('scatter-plot', 'figure'),
    [Input('x-axis-dropdown', 'value'),
     Input('y-axis-dropdown', 'value')]
)
def update_graph(x_axis, y_axis):
    fig = px.scatter(merged_df, x=x_axis, y=y_axis, text='Squad')
    fig.update_traces(textposition='top center')
    fig.update_layout(
        title=f'{y_axis} vs. {x_axis}',
        xaxis_title=x_axis,
        yaxis_title=y_axis,
        height=600,
        width=1000
    )
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
