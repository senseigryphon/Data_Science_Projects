# I recall several observations of Liverpool having 20-25+ shots in a game. I know that they are clear leaders in the Premier League in terms of expected goals and goals scored. I want to see how much of that is because of the volume of shots they take

```python
import pandas as pd
```

```python
df = pd.read_html('https://fbref.com/en/comps/9/Premier-League-Stats')
```

```python
for idx,table in enumerate(df):
 print('***************************')
 print(idx)
 print(table)
```

```python
table_index = 8
table_index_2 = 0
df_shooting = df[table_index]
df_table = df[table_index_2]
```

```python
df_shooting.head()
```

```python
df_table.head()
```

```python
#check the number of levels in the MultiIndex columns
num_levels = df_shooting.columns.nlevels
if num_levels > 1:
    df_shooting.columns = df_shooting.columns.droplevel(0)

#view column names
df_shooting.columns
```

```python
df_shooting['Open Play Shots'] = df_shooting['Sh'] - df_shooting['FK'] - df_shooting['PKatt'] 
```

```python
df_table.columns
```

```python
shooting_stats = df_shooting[['Squad', 'Gls', 'Sh', 'SoT', 'SoT%', 'Sh/90', 'SoT/90',
       'G/Sh', 'G/SoT', 'Dist', 'xG', 'npxG', 'npxG/Sh',
       'G-xG', 'np:G-xG', 'Open Play Shots']]
table = df_table[['Rk', 'Squad', 'Pts', 'Pts/MP', 'xGD', 'xGD/90']]
```

```python
df_merged = pd.merge(shooting_stats, table, on = 'Squad', how = 'inner')
df_merged.head()
```

```python
df_merged.columns
```

```python
rename_dict = {
    'Sh' : 'Shots',
    'Gls' : 'Goals',
    'SoT' : 'Shots on Target',
    'SoT%' : '% of Shots on Target',
    'Sh/90' : 'Shots per 90',
    'SoT/90' : 'Shots on Target per 90',
    'G/Sh' : 'Goals per Shot',
    'G/SoT' : 'Goals per Shot on Target',
    'Dist' : 'Average Shot Distance',
    'xG' : 'Expected Goals',
    'npxG' : 'Non-Penalty Expected Goals',
    'npxG/Sh' : 'Non-Penalty Expected Goals per Shot',
    'G-xG' : 'Expected Goal Performance',
    'np:G-xG' : 'Non-Penalty Expected Goal Performance',
    'Rk' : 'League Position',
    'Pts' : 'Points',
    'Pts/Mp' : 'Points per Match',
    'xGD' : 'Expected Goal Difference',
    'xGD/90' : 'Expected Goal Difference per 90'
    
}

# Use the rename method to change the column names
df_merged = df_merged.rename(columns=rename_dict)
```

```python
import dash
from dash import html, dcc
from dash.dependencies import Input, Output
import plotly.express as px
```

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1('Premier League Shot Data'),
    dcc.Dropdown(
        id='x-axis-dropdown',
        options=[{'label': col, 'value': col} for col in df_merged.columns if col != 'Squad'],
        value='Non-Penalty Expected Goals per Shot',  #default value
        clearable=False,
        style={"width": "50%"}
    ),
    dcc.Dropdown(
        id='y-axis-dropdown',
        options=[{'label': col, 'value': col} for col in df_merged.columns if col != 'Squad'],
        value='Non-Penalty Expected Goals',  #default value
        clearable=False,
        style={"width": "50%"}
    ),
    dcc.Graph(id='scatter-plot')
])

@app.callback(
    Output('scatter-plot', 'figure'),
    [Input('x-axis-dropdown', 'value'),
     Input('y-axis-dropdown', 'value')]
)
def update_graph(x_axis, y_axis):
    fig = px.scatter(df_merged, x=x_axis, y=y_axis)
                     #color='Points',  # Use 'League Position' as a color scale
                     #hover_data=[x_axis, y_axis],
                     #color_continuous_scale=px.colors.sequential.Turbo)
    

    fig.update_traces(marker=dict(size=12, line=dict(width=2, color='DarkSlateGrey')))

    for i, row in df_merged.iterrows():
        fig.add_annotation(
            x=row[x_axis],
            y=row[y_axis],
            text=row['Squad'],
            showarrow=False,
            yshift = 20,
            xshift = 20 # Shift text above the markers
        )

    fig.update_layout(
        title=f'{y_axis} vs. {x_axis} - Premier League Teams',
        xaxis_title=x_axis,
        yaxis_title=y_axis,
        legend_title='Points',
        autosize=True
    )

    fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightPink')
    fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightPink')

    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
